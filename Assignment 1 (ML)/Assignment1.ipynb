```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:/Users/admin/Desktop/Mehzebin/Iris.csv'
iris_df = pd.read_csv(file_path)
iris_df
```

```python
avg_m = iris_df.groupby('Species').mean().drop(columns=['Id'])
```

```python
avg_m.plot(kind='bar', figsize=(9, 5))
plt.title('Bar Plot')
plt.xlabel('Species')
plt.ylabel('Measurements (cm)')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set1')
plt.title('Scatter Plot')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(data=iris_df, x='Species', y='SepalLengthCm', palette='Set3')
plt.title('Box Plot')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```
